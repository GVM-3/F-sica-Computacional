## Segunda derivada numérica de orden 4 con malla uniforme





Mediante el desarrollo de Taylor es posible hacer aproximaciones de la derivada de una función, tales aproximaciones se pueden realizar de forma simétrica y asimétrica; la primera de ellas nos es útil para puntos en el interior de la malla mientras que la aproximación asimétrica nos será de utilidad para los puntos frontera. Es importante notar que como se considera una malla uniforme, la distancia entre el punto $x_{i}$ y los puntos $x_{i-1}, x_{i+1}$ es la misma
    En este caso, se requiere obtener la segunda derivada de una función, de forma sistemática, alrededor del punto $x_i$ con precisión de orden 4, es decir, despreciamos términos en $h⁵$. 

### Puntos en el interior 

Para encontrar dichos puntos partimos de las siguientes ecuaciones:

$$\begin{equation}
f(x_{0}+h) = f(x_{0}) +f'(x_{0})h+\frac{1}{2}f''(x_{0})h^2+\frac{1}{3!}f'''(x_{0})h^3+ \frac{1}{4!}f^{iv}(x_{0})h^4+\mathcal{O⁵} \end{equation}
$$

$$\begin{equation}
f(x_{0}-h) = f(x_{0}) -f'(x_{0})h+\frac{1}{2}f''(x_{0})h^2-\frac{1}{3!}f'''(x_{0})h^3+ \frac{1}{4!}f^{iv}(x_{0})h^4+\mathcal{O⁵} \label{2}\end{equation}
$$

$$\begin{equation}
f(x_{0}+2h) = f(x_{0}) +2f'(x_{0})h+2f''(x_{0})h²+\frac{4}{3}f'''(x_{0})h^3+\frac{2}{3}f^{iv}(x_{0})h^4+\mathcal{O⁵} \label{3}\end{equation}
$$

$$\begin{equation}
f(x_{0}-2h) = f(x_{0}) -2f'(x_{0})h+2f''(x_{0})h²-\frac{4}{3}f'''(x_{0})h^3+\frac{2}{3}f^{iv}(x_{0})h^4+\mathcal{O⁵} \label{4}\end{equation}
$$

Se observa que se tiene un sistema de ecuaciones de 4 ecuaciones con 4 incógnitas, donde la que se desea encontrar es $f''(x_{0})$, de tal forma que sumando la primera ecuación con la segunda y la tercera con la cuarta se obtiene:

$$
f''(x_{0})h^2+ \frac{1}{12}f^{iv}(x_{0})h^4+2\mathcal{O⁵}= f(x_{0}+h) - f(x_{0})+ f(x_{0}-h) - f(x_{0})
$$

$$
4f''(x_{0})h^2+ \frac{16}{12}f^{iv}(x_{0})h^4+2\mathcal{O⁵}= f(x_{0}+2h) - f(x_{0})+ f(x_{0}-2h) + f(x_{0})
$$

Multiplicando por 16 la primera de estas dos últimas ecuaciones y restándosela la segunda a la primera se llega a la siguiente expresión 

$$
12f''(x_{0})h^2= 16f(x_{0}+h)+16f(x_{0}-h)-30f(x_{0})-f(x_{0}+2h)-f(x_{0}-2h)
$$

$$
f''(x_{0})=\frac{\frac{4}{3}f(x_{0}+h)+\frac{4}{3}f(x_{0}-h)-\frac{5}{2}f(x_{0})-\frac{1}{12}f(x_{0}+2h)-\frac{1}{12}f(x_{0}-2h)}{h²}
$$

### Puntos en la frontera

Para calcular la segunda derivada en estos puntos se toman las expresiones para $x_{0}+h, x_{0}+2h, x_{0}+3h, x_{0}+4h$ las cuales forman el siguiente sistema de ecuaciones

$$
f(x_{0}+h) = f(x_{0}) +f'(x_{0})h+\frac{1}{2}f''(x_{0})h^2+\frac{1}{3!}f'''(x_{0})h^3+ \frac{1}{4!}f^{iv}(x_{0})h^4+\mathcal{O⁵} 
$$
$$
f(x_{0}+2h) = f(x_{0})+2f'(x_{0})h+2f''(x_{0})h^2+\frac{4}{3}f'''(x_{0})h^3+ \frac{2}{3}f^{iv}(x_{0})h^4+\mathcal{O⁵} 
$$
$$
f(x_{0}+3h) = f(x_{0}) +3f'(x_{0})h+\frac{9}{2}f''(x_{0})h²+\frac{9}{2}f'''(x_{0})h^3+\frac{27}{8}f^{iv}(x_{0})h^4+\mathcal{O⁵}
$$
$$
f(x_{0}+4h) = f(x_{0}) +4f'(x_{0})h+8f''(x_{0})h²+\frac{32}{3}f'''(x_{0})h^3+\frac{32}{3}f^{iv}(x_{0})h^4+\mathcal{O⁵}
$$


De forma análoga al procedimiento que se utilizó para los puntos en el interior, en el caso de los puntos forntera se determinará la solución de este sistema para $f''(x_{0})$, y de donde se obtiene:


$$
f''(x_0)= \frac{-\frac{26}{3}f(x_{0}+h)+\frac{19}{2}f(x_{0}+2h)-\frac{14}{3}f(x_{0}+3h)+\frac{11}{2}f(x_{0}+4h)-\frac{5}{3}f(x_{0})}{h²}
$$

Puesto que ya se cuenta con las expresiones para obtener la derivada de toda la función se procede a escribir el código. Para ello, se utilizará la biblioteca **Plots** y se usa el siguiente comando:

```julia
using Plots
```

Ahora, considerando que $i=0,1$ del lado izquierdo (los primeros) de los puntos frontera y para los puntos del lado derecho (los últimos) $i=N,N-1$ y se deben realizar los cambios $i+1\rightarrow i-1; i+2 \rightarrow i-2; i+3 \rightarrow i-3; i+4 \rightarrow i-4$ y $h\rightarrow -h$ 

Mediante una función y un ciclo for se relaiza el cálculo de la segunda derivada de una función y se evalúan las expresiones obtenidas anteriormente en cada caso

```julia
function deriv_h1(f,x) #definimos la función que hará el calculo de la derivada  
    f_biprima = zeros(length(f)) 
    h = x[2] - x[1]  
    for i in 3:(length(f)-2)
        #calcular puntos en el interior 
        f_biprima[i] = ((4/3)*f[i+1]+(4/3)*f[i-1]-(1/12)*f[i+2]-(1/12)*f[i-2]-(5/2)*f[i])/(h^2)
  end
    #calcular los primeros últimos puntos 
    f_biprima[1] = ((-26/3)*f[2]+(19/2)*f[3]-(14/3)*f[4]+(11/12)*f[5]+(35/12)*f[1])/(h^2)
    f_biprima[2] = ((-26/3)*f[3]+(19/2)*f[4]-(14/3)*f[5]+(11/12)*f[6]+(35/12)*f[2])/(h^2)
    
    #calcular los dos últimos puntos 
    f_biprima[end] = (-(26/3)*f[end-1]+(19/2)*f[end-2]-(14/3)*f[end-3]+(11/12)*f[end-4]+(35/12)*f[end])/(h^2)
    f_biprima[end-1] = (-(26/3)*f[end-2]+(19/2)*f[end-3]-(14/3)*f[end-4]+(11/12)*f[end-5]+(35/12)*f[end-1])/(h^2)
    
    return f_biprima     
end
```

Para comprobar, graficamos la función con una malla uniforme para la función seno 

```julia
x=collect(range(0,2π,length=500)) #Dominio de la función, que va de 0 a 2π considerando cincuenta puntos dentro de este intervalo 
y = sin.(x) #Función a derivar
y_deriv = -sin.(x) #Segunda derivada analítica 
y_num_deriv = deriv_h1(y,x) # Segunda derivada obtenida numéricamente 
```

Con el comando "plot" se grafican todas las funciones y se obtiene: 

```julia
plot(x,y,label="f(x)")
plot!(x,y_deriv,label="f''(x)")
plot!(x,y_num_deriv,label="f''(x) numérica")
```

## Derivada numérica de orden 2 con malla no uniforme 

A diferencia del cálculo anterior, en este caso se considera una malla no uniforme, es decir, $h_{1}=x_{i+1}-x_{i}, h_{2}=x_{i}-x_{i-1}$ entonces $h_{1}\neq h_{2}$ a diferencia de la malla uniforme donde $h_{1}=h_{2}$ y mediante el desarrollo de taylor se obtiene el siguiente siguiente sistema de ecuaciones de dos ecuaciones y dos incógnitas

### Puntos interiores

$$
f(x_{0}+h_{1}) = f(x_{0}) +f'(x_{0})h_{1}f+\frac{1}{2}f''(x_{0})h^2_{1}
$$

$$
f(x_{0}-h_{2}) = f(x_{0}) -f'(x_{0})h_{2}+\frac{1}{2}f''(x_{0})h^2_{2}
$$

Puesto que se desea encontrar el valor de la primer derivada, se busca encontrar el valor de $f'(x_{0})$, para ello se multiplica la segunda ecuación por $\frac{h²_{1}}{h_²{2}}$ y se obtiene:

$$
-f'(x_{0})\frac{h²_{1}}{h_{2}}+\frac{1}{2}f'(x_{0})h^2_{1}=(f(x_{0}-h_{2})-f(x_{0}))\frac{h²_{1}}{h²_{2}} 
$$

Restando esta última con la primera del sistema de ecuaciones y factorizando los términos comunes llegamos a 

$$
f'(x_{0})\left(h_{1}+\frac{h²_{1}}{h_{2}}\right)=f(x_{0}+h_{1})-f(x_{0})-(f(x_{0}-h_{2})-f(x_{0}))\frac{h²_{1}}{h²_{2}}
$$

$$
f'(x_{0})=\frac{f(x_{0}+h_{1})-f(x_{0})-f(x_{0}-h_{2})\frac{h²_{1}}{h²_{2}}+f(x_{0})\frac{h²_{1}}{h²_{2}}}{h_{1}+\frac{h²_{1}}{h_{2}}}
$$

Finalmente se obtiene que los puntos del interior de la derivada son tales que 


$$
f'(x_{0})=\frac{f(x_{0}+h_{1})h²_{2}-f(x_{0}-h_{2})h²_{1}+f(x_{0})(h²_{1}-h²_{2})}{h_{1}h²_{2}+h²_{1}h_{2}}
$$

### Puntos frontera iniciales

Ahora, para los puntos frontera del inicio se tiene que $h_{1}=x_{i+1}-x_{0}, h_{2}=x_{i+2}-x_{0}$ por lo que el sistema de ecuaciones es el siguiente

$$
f(x_{0}+h_{1}) = f(x_{0}) +f'(x_{0})h_{1}+\frac{1}{2}f''(x_{0})h^2_{1}
$$

$$
f(x_{0}+h_{2}) = f(x_{0}) +f'(x_{0})h_{2}+\frac{1}{2}f''(x_{0})h^2_{2}
$$

Multiplicando la segunda ecuación por $\frac{h²_{1}}{h_²{2}}$ 

$$
f'(x_{0})\frac{h²_{1}}{h_{2}}+\frac{1}{2}f'(x_{0})h^2_{1}=(f(x_{0}+h_{2})-f(x_{0}))\frac{h²_{1}}{h²_{2}} 
$$

Restando este resultado a la primera ecuación del sistema y factorizando términos comunes se llega a que

$$
f'(x_{0})\left(h_{1}-\frac{h²_{1}}{h_{2}}\right)=f(x_{0}+h_{1})-f(x_{0})-(f(x_{0}+h_{2})-f(x_{0}))\frac{h²_{1}}{h²_{2}}
$$

$$
f'(x_{0})=\frac{f(x_{0}+h_{1})-f(x_{0})-f(x_{0}+h_{2})\frac{h²_{1}}{h²_{2}}+f(x_{0})\frac{h²_{1}}{h²_{2}}}{h_{1}-\frac{h²_{1}}{h_{2}}}
$$

La expresión para los primeros puntos de la derivada es:

$$
f'(x_{0})=\frac{f(x_{0}+h_{1})h²_{2}-f(x_{0}+h_{2})h²_{1}+f(x_{0})(h²_{1}-h²_{2})}{h_{1}h²_{2}-h²_{1}h_{2}}
$$

### Puntos frontera finales 


Finalmente, para los últimos puntos frontera se tiene que $h_{1}=x_{N}-x_{N-1}, h_{2}=x_{N}-x_{N-2}$ por lo que el sistema de ecuaciones es el siguiente

$$
f(x_{N}-h_{1}) = f(x_{N}) -f'(x_{N})h_{1}+\frac{1}{2}f''(x_{N})h^2_{1}
$$

$$
f(x_{N}-h_{2}) = f(x_{N}) -f'(x_{N})h_{2}+\frac{1}{2}f''(x_{N})h^2_{2}
$$

Si se multiplica por $\frac{h²_{1}}{h_²{2}}$ la segunda ecuación y se resta con la primera se obtiene:

$$
f'(x_{N})\frac{h²_{1}}{h_{2}}-\frac{1}{2}f'(x_{N})h^2_{1}=f(x_{N}-h_{1})-f(x_{N})-(f(x_{N}-h_{2})-f(x_{N}))\frac{h²_{1}}{h²_{2}} 
$$

$$
f'(x_{N})\left(\frac{h²_{1}}{h_{2}}-h_{1} \right)=f(x_{N}-h_{1})-f(x_{N})-(f(x_{N}-h_{2})-f(x_{N}))\frac{h²_{1}}{h²_{2}}
$$

Y la expresión para los puntos finales de la frontera son:

$$
f'(x_{N})=\frac{f(x_{N}-h_{1})h²_{2}-(f(x_{N}-h_{2})h²_{1}+f(x_{N})(h²_{1}-h²_{2})}{h²_{1}h_{2}-h_{1}h²_{2}}
$$

Una vez obtenidas todas las expresiones para la derivada de segundo orden se continúa escribiendo el código para calcularlas y para ello se define una función con un ciclo for, análogo al caso anterior sólo que en este caso, las distnacias $h_{1}$ y $h_{2}$ varían en cada punto por eso se deben incluir en el ciclo for 

```julia
function deriv_h2(f,x)
    f_prima = zeros(length(f))
    for i in 2:(length(f)-1)
    h1 = x[i+1]-x[i]
    h2 = x[i]-x[i-1]
        
    #puntos en el interior 
    f_prima[i] =((h2^2)*f[i+1]-(h1^2)*f[i-1]+((h1^2)-(h2^2))*f[i])/((h2)*(h1^2)+(h1)*(h2^2))
        
  end
    # primer punto 
    hn1 = x[2]-x[1]
    hn2 = x[3]-x[1]
    f_prima[1]=((hn2^2)f[2]-(hn1^2)f[3]+((hn1^2)-(hn2^2))f[1])/(hn1*(hn2^2)-hn2*(hn1^2))
    
    #último punto
    hnf1 = x[end]-x[end-1]
    hnf2 = x[end]-x[end-2]
    f_prima[end]=((hnf2^2)f[end-1]-(hnf1^2)f[end-2]+((hnf1)^2-(hnf2^2))f[end])/(hnf2*(hnf1^2)-hnf1*(hnf2^2))
    return f_prima
end
```

Para comprobar la función anterior y comprobar con la derivada analítica se realiza el mapeo de algún intervalo y se ajusta al se desee utilizar. En este caso se elevaron los puntos el intervalo $[0,2π]$ a la quinta potencial y posteriormente dividimos por $π^4$. Esto acomoda los puntos en el intervalo de tal manera que éstos se van espaciando como $x^5$.

```julia
x_2=collect(range(0,2π,length=500)) # Dominio de la función
x2 = (x_2.^5)*(1/pi^4)*(1/10) # 
y2 = sin.(x2)
y_deriv2 = cos.(x2)
y_num_deriv2 = deriv_h2(y2,x2)
```

```julia
plot(x2,y2,label="f(x)")
plot!(x2,y_deriv2,label="f'(x)")
plot!(x2,y_num_deriv2,line=:dash,marker=(:dot,2),label="f'(x) numérica")
```

## Segunda derivada númerica de segundo orden con malla no uniforme

Como ya se ha hecho anteriormente, se obtendrá la expresión para la segunda derivada de segundo orden con la misma expresión que el caso anterior (derivada de segundo orden con malla no uniforme), partiendo del siguiente sistema de ecuaciones 

$$
f(x_{0}+h_{1}) = f(x_{0}) +f'(x_{0})h_{1}f+\frac{1}{2}f''(x_{0})h^2_{1}
$$

$$
f(x_{0}-h_{2}) = f(x_{0}) -f'(x_{0})h_{2}+\frac{1}{2}f''(x_{0})h^2_{2}
$$

Que es el mismo que ya se había obtenido previamente, pero en este caso como es de interés obtener la segunda derivada se determinarán las expresiones para $f''(x_{0}$

### Puntos interiores

Se se multiplica la primer ecuación por $\frac{h_{2}}{h_{1}}$ entonces

$$
f'(x_{0})h_{2} +\frac{1}{2}f''(x_{0})h_{2}h_{1}=(f(x_{0}+h_{1})\frac{h_{2}}{h_{1}}-f(x_{0})) 
$$


Sumando este resultado a la segunda ecuación se llega a que 

$$
f''(x_{0})\left(\frac{1}{2}h_{2}h_{1}+\frac{1}{2}h²_{2}\right) =(f(x_{0}+h_{1})-f(x_{0}))\frac{h_{2}}{h_{1}}+f(x_{0}-h_{2})-f(x_{0}) 
$$

$$
f''(x_{0})=\frac{f(x_{0}+h_{1})\frac{h_{2}}{h_{1}}-f(x_{0})\frac{h_{2}}{h_{1}}+f(x_{0}-h_{2})-f(x_{0})}{\frac{1}{2}(h_{2}h_{1}+h²_{2})}
$$

Y las puntos de la segunda derivada quedan determinados por 

$$
f''(x_{0})=\frac{f(x_{0}+h_{1}){h_{2}}+f(x_{0}-h_{2})h_{1}-(h_{2}+h_{1})f(x_{0})}{\frac{1}{2}(h_{2}h²_{1}+h²_{2}h_{1})}
$$

### Puntos frontera iniciales

A partir del siguiente sistema determinaremos los puntos de frontera iniciales, calculando las soluciones para $f'(x_{0})$

$$
f(x_{0}+h_{1}) = f(x_{0}) +f'(x_{0})h_{1}+\frac{1}{2}f''(x_{0})h^2_{1}
$$

$$
f(x_{0}+h_{2}) = f(x_{0}) +f'(x_{0})h_{2}+\frac{1}{2}f''(x_{0})h^2_{2}
$$

Cambiamos la primera de las dos ecuaciones ecuación por: 

$$
f'(x_{0})h_{2}+\frac{1}{2}f''(x_{0})h_{1}h_{2}=(f(x_{0}+h_{1})-f(x_{0}))\frac{h_{2}}{h_{1}}
$$

Restando esta última con la primera del sistema se encuentra que

$$
f''(x_{0})\left(\frac{1}{2}h_{2}h_{1}-\frac{1}{2}h^2_{2}\right)=(f(x_{0}+h_{1})-f(x_{0}))\frac{h_{2}}{h_{1}}-(f(x_{0}+h_{2})-f(x_{0})
$$

$$
f''(x_{0})=\frac{f(x_{0}+h_{1})h_{2}-f(x_{0}+h_{2})h_{1}-f(x_{0})(h_{2}-h_{1})}{\frac{1}{2}(h_{2}h²_{1}-h_{1}h²_{2})}
$$

Es la expresión para los puntos frontera iniciales.

### Puntos frontera final 

En este caso, se utiliza el siguiente sistema de ecuaciones (planteado anteriormente) para encontrar los puntos últimos puntos frontera de $f''(x_{N})$

$$
f(x_{N}-h_{1}) = f(x_{N}) -f'(x_{N})h_{1}+\frac{1}{2}f''(x_{N})h^2_{1}
$$

$$
f(x_{N}-h_{2}) = f(x_{N}) -f'(x_{N})h_{2}+\frac{1}{2}f''(x_{N})h^2_{2}
$$

Tomando la primera de estas dos expresiones y multiplicandola por el factor $\frac{h_{2}}{h_{1}}$ 

$$
\frac{1}{2}f''(x_{N})h_{2}-f'(x_{N})h^2_{1}=f(x_{N}-h_{1})-f(x_{N})\frac{h_{1}}{h_{2}} 
$$

Restanle esta última a la primera ecuación:

$$
f''(x_{N})\left(\frac{1}{2}h_{2}h_{1}-\frac{1}{2}h^2_{2}\right)=(f(x_{N}-h_{1})-f(x_{N}))\frac{h_{2}}{h_{1}}-(f(x_{N}-h_{2})-f(x_{N})
$$

$$
f''(x_{N})=\frac{f(x_{N}-h_{1})h_{2}-f(x_{N}-h_{2})h_{1}-f(x_{N})(h_{2}-h_{1})}{\frac{1}{2}(h_{2}h²_{1}-h_{1}h²_{2})}
$$

He aquí la expresión para los últimos puntos de la frontera de la segunda derivada de segundo orden con malla no uniforme 

```julia
function deriv_h3(f,x)
    f_biprima2 = zeros(length(f))
    for i in 2:(length(f)-1)
    h1 = x[i+1]-x[i]
    h2 = x[i]-x[i-1]
    #puntos en el interior 
    f_biprima2[i] =(2*(h2*f[i+1]+h1*f[i-1]-(h1^2+h2^2)*f[i]))/(h2*(h1^2)+h1*(h2^2))
        
  end
      # primer punto 
    hm1 = x[2]-x[1]
    hm2 = x[3]-x[1]
    f_biprima2[1]=(2*((hm2)f[2]-(hm1)f[3]-(hm2-hm1)f[1]))/(hm2*(hm1^2)-hm1*(hm2^2))
    
        #último punto
    hmf1 = x[end]-x[end-1]
    hmf2 = x[end]-x[end-2]
    f_biprima2[end]=(2*((hmf2)f[end-1]-(hmf1)f[end-2]-((hmf2)-(hmf1))f[end]))/(hmf2*(hmf1^2)-hmf1*(hmf2^2))
    return f_biprima2
end
```

```julia
x_3=collect(range(0,2π,length=500)) # Dominio de la función
x3 = (x_2.^5)*(1/pi^4)*(1/10) # 
y3 = sin.(x3)
y_deriv3 = cos.(x3)
y_num_deriv3 = deriv_h3(y3,x3)
```

# Errores entre la solución analítica y la numérica para cada aproximación 

Consideremos una función $f$ tal que 

$$
f(x)=e^{-x^2}sen(10x) \space\space x\epsilon[-5,5]$$ 

Entonces 

$$
f'(x)=e^{-x²}[10cos(10x)-2xsen(10x))]
$$

$$
f''(x)=e^{-x²}[(4x²-102)sen(10x)-40xcos(10x)]
$$

# Experimento numérico

Ahora, si consideramos la función 


$$
f(x)=\frac{sen(3x)}{3x} \space\space x\epsilon[-2π,2π] 
$$

Y su derivada es

$$
f'(x)=\frac{cos(3x)}{x}-\frac{sen(3x)}{3x²}
$$

Anteriormente se obtuvieron expresiones para la derivada con una malla uniforme, retomando el sistema de ecuaciones que se utilizó en ese caso tenemos que:

$$
f'(x_{0})h+\frac{1}{2}f''(x_{0})h²=f(x_{0}+h)-f(x_{0})
$$

$$
-f'(x_{0})h+\frac{1}{2}f''(x_{0})h²=f(x_{0}-h)-f(x_{0})
$$

Entonces
$$
f'(x_{0})=\frac{f(x_{0}+h)-f(x_{0}-h)}{2h}
$$


Este resultado es para los puntos interiores.

Ahora, para los puntos frontera se tiene que 

$$
f'(x_{0})h+\frac{1}{2}f''(x_{0})h²=f(x_{0}+h)-f(x_{0})
$$

$$
2f'(x_{0})h+2f''(x_{0})h²=f(x_{0}-h)-f(x_{0})
$$


Los puntos frontera son:
$$
f'(x_{0})=\frac{2f(x_{0}+h)-\frac{1}{2}f(x_{0}+2h)+\frac{3}{2}f(x_{0})}{h}
$$


El código para realizar esta operación será similar a como se ha venido realizando y quedaría de la siguiente forma:

```julia
function deriv_h5(f,x)          #definimos la función que hará el calculo de la derivada 
    f_prima5 = zeros(length(f))
    h = x[2] - x[1]
    dosh = 2*h
    for i in 2:(length(f)-1)         #calcular puntos en el interior 
        f_prima5[i] = ( f[i+1]-f[i-1] )/(dosh)
    end
    # calcular el primer punto
    f_prima5[1] = ( -0.5*f[3] + 2*f[2] -1.5*f[1] )/h
    # calcular el ultimo punto
    f_prima5[end] = -( -0.5*f[end-2] + 2*f[end-1] -1.5*f[end] )/h    
    return f_prima5
end
```

```julia
x5=collect(range(-2π,2π,length=500)) #Dominio de la función 
y5= zeros(length(x5))       
x_5= x5.^2
y_deriv5=zeros(length(x5)) 
for i in 1:length(x5)
   y5[i] = (sin.(3*x5[i]))/(3*x5[i])  #Función original
    y_deriv5[i]= cos.(x5[i])/x5[i]-(sin.(3*x5[i]))/(x5[i])     #Primera derivada
    
end

y_num_deriv5= deriv_h5(y5,x5) 
```

```julia
plot(x5,y5,label="f(x)")
plot!(x5,y_deriv5,label="f''(x)")
plot!(x5,y_num_deriv5,label="f''(x) numérica")
```

Para la derivada de **cuarto orden** se obtiene del mismo sistema que el primer ejercicio y como resultado se obtiene que los puntos interiores son

$$
f'(x_{0})=\frac{\frac{2}{3}f(x_{0}+h)-\frac{2}{3}f(x_{0}-h)-\frac{1}{12}f(x_{0}+2h)+\frac{1}{12}f(x_{0}-2h)}{h}
$$

Por su parte, los puntos en la frontera están determinados por
$$
f'(x_{0})=\frac{4f(x_{0}+h)-3f(x_{0}+2h)+\frac{4}{3}f(x_{0}+3h)-\frac{1}{4}f(x_{0}+4h)+\frac{25}{12}f(x_{0})}{h}
$$

Así mismo, se escribe el código con las operaciones necesarias para realizar la derivada como se muestra a continuación

```julia
function deriv_h4(f,x) #definimos la función que hará el calculo de la derivada  
    f_primaf = zeros(length(f)) 
    h = x[2] - x[1]  
    for i in 3:(length(f)-2)
        #calcular puntos en el interior 
        f_primaf[i] = ((2/3)*f[i+1]-(2/3)*f[i-1]-(1/12)*f[i+2]+(1/12)*f[i-2])/h
  end
    #calcular los primeros últimos puntos 
    f_primaf[1] = (4*f[2]-3*f[3]+(4/3)*f[4]-(1/4)*f[5]-(25/12)*f[1])/h
    f_primaf[2] =  (4*f[3]-3*f[4]+(4/3)*f[5]-(1/4)*f[6]-(25/12)*f[2])/h
    #calcular los dos últimos puntos 
    f_primaf[end]   = (4*f[end-1]-3*f[end-2]+(4/3)*f[end-3]-(1/4)*f[end-4]-(25/12)*f[end])/-h
    f_primaf[end-1] = (4*f[end-2]-3*f[end-3]+(4/3)*f[end-4]-(1/4)*f[end-5]-(25/12)*f[end-1])/-h
    
    return f_primaf     
end
```
